```python
import re
from collections import Counter

def conta_termos(texto, termos_dict):
    contador = Counter()
    for termo in termos_dict:
        ocorrencias = len(re.findall(r'\b{}\b'.format(re.escape(termo)), str(texto).lower()))
        if ocorrencias > 0:
            contador[termo] += ocorrencias
    return contador

contador_geral = Counter()

# Ajuste a coluna aqui de acordo com o nome correto
for especificacao in df['especificacao']:
    contador_geral.update(conta_termos(especificacao, termos_dict))

# Obter os 10 termos mais frequentes
top_10_termos = contador_geral.most_common(10)

# Exibir os resultados
print("Os 10 termos mais frequentes são:")
for termo, frequencia in top_10_termos:
    print(f"{termo}: {frequencia}")

# Salvar os resultados em um novo Excel, se necessário
top_10_df = pd.DataFrame(top_10_termos, columns=['Termo', 'Frequencia'])
output_path = '/content/drive/MyDrive/resultado_top_10_termos.xlsx'
top_10_df.to_excel(output_path, index=False)
```

```python
import re
from collections import Counter, defaultdict

def conta_termos_e_recupera_linhas(texto, termos_dict, pergunta, resposta):
    contador = Counter()
    linhas_associadas = defaultdict(list)
    for termo in termos_dict:
        ocorrencias = len(re.findall(r'\b{}\b'.format(re.escape(termo)), str(texto).lower()))
        if ocorrencias > 0:
            contador[termo] += ocorrencias
            linhas_associadas[termo].append((texto, pergunta, resposta))
    return contador, linhas_associadas

contador_geral = Counter()
linhas_geral = defaultdict(list)

for _, row in df.iterrows():
    especificacao = row['especificacao']
    pergunta = row['pergunta']
    resposta = row['resposta']
    contador, linhas_associadas = conta_termos_e_recupera_linhas(especificacao, termos_dict, pergunta, resposta)
    contador_geral.update(contador)
    for termo, linhas in linhas_associadas.items():
        linhas_geral[termo].extend(linhas)
```

```python
# Montar o Google Drive
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# Carregar o dicionário de termos
dict_path = '/content/drive/MyDrive/Documentos-stt/Dicionário-de-termos/dic-cid10/__Dicionário Único de Termos.txt'
termos_dict = {}

with open(dict_path, 'r', encoding='utf-8') as file:
    for line in file:
        term, freq = line.strip().split(': ')
        termos_dict[term.strip()] = int(freq)

# Carregar o arquivo de teleconsultorias
teleconsultorias_path = '/content/drive/MyDrive/Documentos-stt/Teleconsultorias-anonimizadas-v1 - Teleconsultorias-anonimizadas-v1 (1).csv'
df = pd.read_csv(teleconsultorias_path)

# Verificar as colunas do DataFrame
print(df.columns)

# Importar e configurar o FlashText
!pip install flashtext
from flashtext import KeywordProcessor
from collections import Counter, defaultdict

# Configurar o KeywordProcessor
keyword_processor = KeywordProcessor()
for term in termos_dict:
    keyword_processor.add_keyword(term)

def conta_termos(texto, keyword_processor):
    termos_encontrados = keyword_processor.extract_keywords(str(texto).lower())
    return Counter(termos_encontrados)

contador_geral = Counter()
linhas_geral = defaultdict(list)

# Ajuste a coluna aqui de acordo com o nome correto
for _, row in df.iterrows():
    especificacao = row['especificacao']
    pergunta = row['pergunta']
    resposta = row['resposta']
    termos_encontrados = keyword_processor.extract_keywords(str(especificacao).lower())
    for termo in termos_encontrados:
        contador_geral[termo] += 1
        linhas_geral[termo].append((especificacao, pergunta, resposta))

# Obter os 10 termos mais frequentes
top_10_termos = contador_geral.most_common(10)

# Exibir os resultados
print("Os 10 termos mais frequentes são:")
for termo, frequencia in top_10_termos:
    print(f"{termo}: {frequencia}")

# Criar um DataFrame para salvar os resultados no Excel
dados = []
linhas_count = 0
for termo, _ in top_10_termos:
    for especificacao, pergunta, resposta in linhas_geral[termo]:
        if linhas_count < 10:
            dados.append([termo, especificacao, pergunta, resposta])
            linhas_count += 1
        else:
            break

top_10_df = pd.DataFrame(dados, columns=['Termo', 'Especificacao', 'Pergunta', 'Resposta'])

# Salvar os resultados em um novo Excel
output_path = '/content/drive/MyDrive/resultado_top_10_termos_com_perguntas_respostas.xlsx'
top_10_df.to_excel(output_path, index=False)

# Mostrar o DataFrame
top_10_df.head(10)
```
